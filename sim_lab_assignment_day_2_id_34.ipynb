2. Analyze Matrix A
Using NumPy, compute:
● Shape of A
● Determinant
● Rank
● Eigenvalues
● Inverse (only if determinant ≠ 0)

```python
A = np.array([[6, 4], [8, 5]])
size_A = A.shape
det_A = np.linalg.det(A)
rank_A = np.linalg.matrix_rank(A)
eigenvalues= np.linalg.eig(A)
if np.linalg.det(A) != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "Matrix is Singular, cannot compute"

print("\nsize_A:\n", size_A)
print("\ndet_A:\n", det_A)
print("\nrank_A:\n", rank_A)
print("\neigenvalues:\n", eigenvalues)
print("\ninv_A:\n", inv_A)
```
